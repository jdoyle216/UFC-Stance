```python
df1 = df[df.R_Stance != 'Open Stance']
df1 = df1[df.B_Stance != 'Open Stance']
```

```python
df1
```

```python
conditions = [
    (df1['R_Stance'] == 'Orthodox') & (df1['B_Stance'] == 'Orthodox'),
    (df1['R_Stance'] == 'Orthodox') & (df1['B_Stance'] == 'Southpaw'),
    (df1['R_Stance'] == 'Southpaw') & (df1['B_Stance'] == 'Southpaw'),
    (df1['R_Stance'] == 'Southpaw') & (df1['B_Stance'] == 'Orthodox'),
    (df1['R_Stance'] == 'Orthodox') & (df1['B_Stance'] == 'Switch'),
    (df1['R_Stance'] == 'Switch') & (df1['B_Stance'] == 'Orthodox'),
    (df1['R_Stance'] == 'Southpaw') & (df1['B_Stance'] == 'Switch'),
    (df1['R_Stance'] == 'Switch') & (df1['B_Stance'] == 'Southpaw'),
    (df1['R_Stance'] == 'Switch') & (df1['B_Stance'] == 'Switch')
    
    ]

# create a list of the values we want to assign for each condition
values = ['Orth_vs_Orth','Ortho_vs_Southpaw','Southpaw_vs_Southpaw','Ortho_vs_Southpaw','Ortho_vs_Switch','Ortho_vs_Switch','Southpaw_vs_Switch','Southpaw_vs_Switch','Switch_vs_Switch']

# create a new column and use np.select to assign values to it using our lists as arguments
df1['Stances'] = np.select(conditions, values)

# display updated DataFrame
df1.head()
```
